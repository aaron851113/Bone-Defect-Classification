```python
#匯入運作所需套件
# -*- coding: utf-8 -*-
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
```

```python
#確認是否可用 GPU；若否，則使用 CPU。
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
print('GPU state:', device)
```

若上述執行完，結果為 GPU state : cpu 代表沒使用到GPU <br/>
執行以下動作修改 <br/>
- 執行階段
- 變更執行階段類型
- 更改成gpu 

```python
#再次使用 torchvision 內的 transforms 來進行圖片的轉換。這真的是個非常好用的功能，我最近也在寫篇文章紀錄各種功能。#Normaliz正規化
# Cifar-10 data
#transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
#由於 PyTorch 的 datasets 裡面便有附 CIFAR-10，於是這裡便可以直接載下來，不用再另外手動設定。

#如果沒有 data 資料夾存於當前目錄，會自動創建一個資料夾並將 CIFAR-10 的資料放在裡面。

#另外，batch_size 其實可以自己調整，不過我目前嘗試 Accuracy 最高的便是 8 了。
# Data
#trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
#testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

traindir = './defect/Train/'
testdir = './defect/Test/'
 
trainset = torchvision.datasets.ImageFolder( traindir,
        transforms.Compose([
            transforms.RandomResizedCrop(224),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            normalize,
        ]))

testset = torchvision.datasets.ImageFolder( testdir,
        transforms.Compose([
            transforms.RandomResizedCrop(224),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            normalize,
        ]))

trainLoader = torch.utils.data.DataLoader(trainset, batch_size=256, shuffle=True, num_workers=8) #batch8代表一次只丟8張圖進去訓練
testLoader = torch.utils.data.DataLoader(testset, batch_size=128, shuffle=False, num_workers=8)
```

```python
#CIFAR-10 當中的 10 種分類。

# Data classes
classes = ('Fine','Broke')
```

```python
#model架構(之後會更換成其他影像分類的cnn model
# Model structure
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 5, 2)
        self.conv2 = nn.Conv2d(16, 32, 5, 2)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32*13*13, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 2)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 32*13*13)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

net = Net().to(device)
print(net)
```

```python
#參數設定。分別是 Loss function (採用多分類器經典的 CrossEntropy) 、Learning Rate、迭代次數、優化器。
# Parameters
criterion = nn.CrossEntropyLoss()#設定如何計算分類錯誤率(後續會藉由這個錯誤率，來修改model架構的權重)
lr = 0.001
epochs = 10
optimizer = optim.SGD(net.parameters(), lr=lr, momentum=0.9)
```

```python
#這裡便是訓練的過程了。要注意的是記得 optimizer.zero_grad() 要每次在更新權重前清空梯度，不然梯度會一直累積。
# Train(train的目的是把圖片丟進model，得到錯誤率，藉由錯誤率來更新model的權重)
for epoch in range(epochs): #epochs代表要訓練幾次，美訓練一次代表把所有圖丟進model跑一次
    running_loss = 0.0

    for times, data in enumerate(trainLoader, 0):
        inputs, labels = data #從data取出圖片(inputs)跟標註(labels)
        inputs, labels = inputs.to(device), labels.to(device) #把圖片跟鰾住丟進gpu，才能進行gpu加速
        # Zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs) #outputs是把圖片丟進model(net)後得到的結果   (output可能是對的可能是錯的)
        loss = criterion(outputs, labels) #把outputs跟標籤去做比對，錯誤率的計算方法是crossentropy
        loss.backward() #loss就是這一次訓練得到的錯誤率，做backward(反向傳播)，就是在更新model權重
        optimizer.step()

        # print statistics
        running_loss += loss.item()

        if times % 1000 == 999 or times+1 == len(trainLoader):
            print('[%d/%d, %d/%d] loss: %.3f' % (epoch+1, epochs, times+1, len(trainLoader), running_loss/2000))

print('Finished Training') 

# 訓練完後會得到一組最適合的權重
```

```python
# Test 藉由training得到最適合的權重後，要拿來做測試
correct = 0
total = 0
with torch.no_grad():
    for data in testLoader:
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = net(inputs) # ex: output = [ 0.1, 0.15, 0.05, .. 0.7, 0.001 ]
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on test inputs: %d %%' % (100 * correct / total))

class_correct = list(0. for i in range(len(classes)))
class_total = list(0. for i in range(len(classes)))
with torch.no_grad():
    for data in testLoader:
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = net(inputs)
        _, predicted = torch.max(outputs, 1)
        c = (predicted == labels).squeeze()
        for i in range(len(classes)):
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1

for i in range(len(classes)):
    print('Accuracy of %5s : %2d %%' % (classes[i], 100 * class_correct[i] / class_total[i]))
```
