```python
#https://chih-sheng-huang821.medium.com/03-pytorch-dataaug-a712a7a7f55e

import PIL.Image as Image
import torchvision
import matplotlib.pyplot as plt
import numpy as np
import warnings
import glob
from torchvision import transforms
from torchvision.transforms import functional as TF
warnings.filterwarnings("ignore")
```

### Test data augmentation

```python
# read image with PIL module
img_pil = Image.open('./defect/Train/Broke/L01-P0000.tif', mode='r')
img_pil = img_pil.convert('RGB')
print(img_pil.size)
plt.imshow(img_pil)
```

```python
#random crop
size=(1200, 1200)
transform = transforms.Compose([
    transforms.RandomCrop(size)
])
new_img = transform(img_pil)
plt.imshow(new_img)
```

```python
# RandomHorizontalFlip
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=1),
])
new_img = transform(img_pil)
plt.imshow(new_img)
```

```python
# RandomVerticalFlip
transform = transforms.Compose([
    transforms.RandomVerticalFlip(p=1),
])

new_img = transform(img_pil)
plt.imshow(new_img)
```

```python
transform = transforms.Compose([
    transforms.RandomAffine(degrees=(-45,45), shear=(0,0), fillcolor=(255,255,255))
])

new_img = transform(img_pil)
plt.imshow(new_img)
```

### implement data augmentation

```python
image_list = sorted(glob.glob('./defect/Train/Broke/*'))
augmentation_folder = './defect/Augmentation/'

for i, imagepath in enumerate(image_list):
    img_pil = Image.open(imagepath, mode='r')
    img_pil = img_pil.convert('RGB')
    #print(img_pil.size)
    
    # RandomHorizontalFlip
    transform = transforms.Compose([transforms.RandomHorizontalFlip(p=1)])
    
    HorizontalFlip_img = transform(img_pil)
    img_name = augmentation_folder + imagepath.split('/')[-1][:-4]+ '_HorizontalFlip' + '.tif'
    HorizontalFlip_img.save(img_name)
    
    # RandomVerticalFlip
    transform = transforms.Compose([transforms.RandomVerticalFlip(p=1)])
    
    VerticalFlip_img = transform(img_pil)
    img_name = augmentation_folder + imagepath.split('/')[-1][:-4]+ '_VerticalFlip' + '.tif'
    VerticalFlip_img.save(img_name)
    
    # RandomAffine
    transform = transforms.Compose([transforms.RandomAffine(degrees=(-30,30), shear=(0,0), fillcolor=(255,255,255))])
    
    RandomAffine_img1 = transform(img_pil)
    img_name = augmentation_folder + imagepath.split('/')[-1][:-4]+ '_RandomAffine1' + '.tif'
    RandomAffine_img1.save(img_name)
    
    RandomAffine_img2 = transform(img_pil)
    img_name = augmentation_folder + imagepath.split('/')[-1][:-4]+ '_RandomAffine2' + '.tif'
    RandomAffine_img2.save(img_name)
    
    if i % 100 == 0:
        print(i, end=' ')

    
```
